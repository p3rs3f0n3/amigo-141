Este notebook explora los datasets principales de datos.gov.co:
- Saber 3°, 5° y 9° (`4h84-62xp`)
- MEN estadísticas por municipio (`nudc-7mev`)
- DIVIPOLA geolocalizados (`vafm-j2df`)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_colwidth', 120)

BASE_URL = 'https://www.datos.gov.co/resource'
```

```python
# Cargar Saber 3°, 5° y 9°
url_saber = f'{BASE_URL}/4h84-62xp.csv?$limit=5000'
df_saber = pd.read_csv(url_saber, dtype=str, low_memory=False)
print(f'Saber 3-5-9: {len(df_saber):,} filas | {len(df_saber.columns)} columnas')
df_saber.head()
```

```python
# Explorar columnas disponibles
print('Columnas Saber 3-5-9:')
for col in df_saber.columns:
    nulos = df_saber[col].isna().sum() + (df_saber[col] == '').sum()
    print(f'  {col:<50} nulos: {nulos:,}')
```

```python
# Cargar MEN municipio
url_men = f'{BASE_URL}/nudc-7mev.csv?$limit=5000'
df_men = pd.read_csv(url_men, dtype=str, low_memory=False)
print(f'MEN municipio: {len(df_men):,} filas | {len(df_men.columns)} columnas')
df_men.head()
```

```python
# Distribución de municipios en Saber
if 'cole_mcpio_ubicacion' in df_saber.columns:
    print('Top 10 municipios en Saber 3-5-9:')
    print(df_saber['cole_mcpio_ubicacion'].value_counts().head(10))
```
